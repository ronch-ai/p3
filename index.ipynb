```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

import pyecharts.options as opts
from pyecharts.charts import Line

xaxis_data = ['1-24', '1-25', '1-26', '1-27', '1-28', '1-29', '1-30', '1-31']
yaxis_data_a = [897, 1408, 2076, 2857, 4630, 6095, 8149]
yaxis_data_b = [1076, 2032, 2692, 5794, 6973, 9239, 12167]


def line_base() -> Line:
    c = (
        Line()
        .add_xaxis(xaxis_data)
        .add_yaxis("确诊人数", yaxis_data_a, is_symbol_show=True, symbol='rect')
        .add_yaxis("疑似感染", yaxis_data_b)
        .set_global_opts(title_opts=opts.TitleOpts(title="全国疫情趋势图"))
    )
    return c

result = line_base()
result.load_javascript()
```

```python
result.render_notebook()
```

```python
import time
import requests
r = requests.get('https://lab.isaaclin.cn/nCoV/api/overall?latest=0')
results = r.json()['results']
for result in results:
    timeArray = time.localtime(result['updateTime']/1000)
    readableTime = time.strftime("%Y-%m-%d %H:%M:%S", timeArray)
    confirmCount = result['confirmedCount']
    suspectedCount = result['suspectedCount']
    curedCount = result['curedCount']
    deadCount = result['deadCount']
    print("更新时间: ", readableTime, ' 确诊人数: ', confirmCount, ' 疑似病例: ', suspectedCount, ' 治愈人数: ', curedCount, ' 死亡人数: ', deadCount)
```

```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

from pyecharts.faker import  Faker
from pyecharts import options as opts
from pyecharts.charts import Map

def map_visualmap() -> Map:
    c = (
        Map()
        .add("确诊人数", [list(z) for z in zip(Faker.provinces, Faker.values())], "china")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Map-VisualMap（分段型）"),
            visualmap_opts=opts.VisualMapOpts(max_=200, is_piecewise=True),
        )
    )
    return c
resultMap = map_visualmap()
resultMap.load_javascript()
```

```python
resultMap.render_notebook()
```

```python
import json
import os

from pyecharts import options as opts
from pyecharts.charts import Page, PictorialBar
from pyecharts.globals import SymbolType

import time
import requests
r = requests.get('https://lab.isaaclin.cn/nCoV/api/area?latest=1&province=湖北省')
results = r.json()['results']
citiesName = []
confirmCount = []

cities = results[0]['cities']
for city in cities:
    citiesName.append(city['cityName'])
    confirmCount.append(city['confirmedCount'])

def pictorialbar_base() -> PictorialBar:
    c = (
        PictorialBar()
        .add_xaxis(citiesName)
        .add_yaxis(
            "",
            confirmCount,
            label_opts=opts.LabelOpts(is_show=False),
            symbol_size=18,
            symbol_repeat="fixed",
            symbol_offset=[0, 0],
            is_symbol_clip=True,
            symbol=SymbolType.ROUND_RECT,
        )
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts(title="福建省疫情柱状图-数据来源网络"),
            xaxis_opts=opts.AxisOpts(is_show=False),
            yaxis_opts=opts.AxisOpts(
                axistick_opts=opts.AxisTickOpts(is_show=False),
                axisline_opts=opts.AxisLineOpts(
                    linestyle_opts=opts.LineStyleOpts(opacity=0)
                ),
            ),
        )
    )
    return c
resultBar = pictorialbar_base()
resultBar.load_javascript()
```

```python
resultBar.render_notebook()
```

```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

import json
import os

from pyecharts import options as opts
from pyecharts.charts import Page, PictorialBar
from pyecharts.globals import SymbolType

import requests

appid = '37734679'
appsecret = 'vfHs0Wf4'
requrl = 'https://tianqiapi.com/api?version=epidemic&appid=37734679&appsecret=vfHs0Wf4'
provinceName = '福建'

citiesName = []
confirmedCount = []

r = requests.get(requrl)
result = r.json()
area = result['data']['area']
for province in area:
    if province['provinceName'] == provinceName:
        cities = province['cities']
        for city in cities:
            citiesName.append(city['cityName'])
            confirmedCount.append(city['confirmedCount'])

def pictorialbar_base() -> PictorialBar:
    c = (
        PictorialBar()
        .add_xaxis(citiesName)
        .add_yaxis(
            "确诊人数",
            confirmedCount,
            label_opts=opts.LabelOpts(is_show=False),
            symbol_size=18,
            symbol_repeat="fixed",
            symbol_offset=[0, 0],
            is_symbol_clip=True,
            symbol=SymbolType.ROUND_RECT,
        )
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts(title=provinceName + "疫情柱状图-数据来源网络"),
            xaxis_opts=opts.AxisOpts(is_show=False),
            yaxis_opts=opts.AxisOpts(
                axistick_opts=opts.AxisTickOpts(is_show=False),
                axisline_opts=opts.AxisLineOpts(
                    linestyle_opts=opts.LineStyleOpts(opacity=0)
                ),
            ),
        )
    )
    return c
cityPictorialBar = pictorialbar_base()
cityPictorialBar.load_javascript()
```

```python
cityPictorialBar.render_notebook()
```
